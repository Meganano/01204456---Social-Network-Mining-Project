```python
import pandas as pd
import json
import sys
import ast
from apyori import apriori
```

```python
path = 'C:/Users/Dell/Desktop/Year 3/Semester2/Social Net mining/Final project/ptlog_20181015.json'
```

```python
datas = []
count = 0
file = open(path,encoding="utf8")

for line in file:
    try:
        if (line == '\n'):      # skip empty lines
            continue
        count = count + 1
        #tweet = json.loads(line)
        data = ast.literal_eval(line)
        datas.append(data)
        if count%10000 == 0:
            sys.stdout.write('.')
        if count%100000 == 0:
            sys.stdout.write('\n')
        if count%100000 == 0:
            break
    except Exception as e:
        print(e)
        continue
print(f"\n{count} datas read.")
```

```python
data = pd.DataFrame.from_dict(datas)
data
```

```python
df = data[['tc', 'topic_id']]
df
```

```python
#อันนี้แค่ลองเทสเฉยๆว่ามันจิ้มหาตัวที่มี tc ซ้ำกันได้ไหม
# ไม่ต้องสนใจก็ได้
test = df.loc[df['tc'] == 'pep2gd5biFyoysO3PlC']
test
```

```python
# ไว้ตรวจสอบจำนวนการซ้ำของ tc
# ไม่ต้องสนใจก็ได้
gb = df.groupby('tc', sort = False)
gb_count = gb.count()
gb_count['topic_id']
```

```python
## Eliminate duplicate tc
df_no_du = df.drop_duplicates(subset ='tc',keep='first',inplace =False)

length = df_no_du.shape[0]+1
sum_tc = 0
tc_no_du = []

#ดูว่ามี tc อะไรบ้างแล้วเอาใส่ list tc_no_du เพื่อเอาไปใช้เช็คต่อว่า tc คนนี้เข้าtopic_idอะไรบ้าง
for i in range(0,length):
    try:
        print(df_no_du['tc'][i])
        tc_no_du.append(df_no_du['tc'][i])
        sum_tc = sum_tc +1
    except:
        continue
```

```python
#เช็คว่าจำนวนtcในlist(tc_no_du) == จำนวนที่วิ่งนับในลูป(sum_tc)
print("sum_tc: ",sum_tc)
print("length of tc_no_du ",len(tc_no_du),"\n")
```

```python
# เช็คว่า tc เข้า topic_id อะไรแล้วใส่ใน records
records = []
count = 0
for i in range(0,sum_tc):
    try:
        tmp = df.loc[df['tc'] == tc_no_du[i]]
        records.append(str(j) for j in tmp['topic_id'])
    except:
        print("\n error at count = ", count ,'\n')
        print(tmp)
    count = count + 1
    if count%100 == 0:
            sys.stdout.write('.')
    if count%10000 == 0:
            sys.stdout.write('\n')        

print("\nread ", count)
```

```python
#หลังจากนี้คือพังหรือเปล่าไม่รู้เย้
association_rules = apriori(records, min_support=0.0000045,min_length=2)
association_results = list(association_rules)  
association_results
```
